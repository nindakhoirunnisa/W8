```python
from sklearn.model_selection import train_test_split

X = df.drop(['District_Code', 'District', 'Life_Expectancy'], axis=1).values
y = df['Life_Expectancy'].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, random_state = 42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
sklearn_model= LinearRegression().fit(X_train, y_train)
sklearn_y_predictions= sklearn_model.predict(X_train)
sklearn_y_predictions
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error 
mean_absolute_error(sklearn_y_predictions, y_train)
```

MAE refers to average of difference between true value and prediction. Since, it is 0.0267 means the predicted values are close to true values.

```python
mean_squared_error(sklearn_y_predictions, y_train)
```

```python
y_pred = sklearn_model.predict(X_test)
y_pred
```

```python
mean_absolute_error(y_pred, y_test)
```

```python
mean_squared_error(y_pred, y_test)
```

```python
preds = sklearn_model.predict(X)
```

MSE measures amount of error in the statistical model. MSE value is near zero and is non-negative. Therefore, good fit.
